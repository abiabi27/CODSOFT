```python
import tkinter as tk
from tkinter import scrolledtext


qa_pairs = {
    "What is your name?": "My name is ChatBot.",
    "How are you?": "I'm just a computer program, so I don't have feelings, but I'm here to help you.",
    "What can you do?": "I can answer questions and have simple conversations with you.",
    "Goodbye": "Goodbye! Feel free to ask more questions anytime.",
}


def send_message():
    user_message = user_input.get()
    if user_message.strip() == "":
        return
    user_input.delete(0, tk.END)
    chat_display.config(state=tk.NORMAL)
    chat_display.insert(tk.END, "You: " + user_message + "\n")
    chat_display.config(state=tk.DISABLED)
    
    if user_message in qa_pairs:
       
        bot_response = qa_pairs[user_message]
    else:
       
        bot_response = qa_pairs.get(user_message, "I don't know the answer to that question.")
    
    chat_display.config(state=tk.NORMAL)
    chat_display.insert(tk.END, "Bot: " + bot_response + "\n")
    chat_display.config(state=tk.DISABLED)
    chat_display.yview(tk.END)


def add_custom_qa():
    custom_question = custom_question_entry.get()
    custom_answer = custom_answer_entry.get()
    if custom_question.strip() == "" or custom_answer.strip() == "":
        return
    qa_pairs[custom_question] = custom_answer
    custom_question_entry.delete(0, tk.END)
    custom_answer_entry.delete(0, tk.END)


window = tk.Tk()
window.title("ChatBot")


chat_display = scrolledtext.ScrolledText(window, wrap=tk.WORD, state=tk.DISABLED)
chat_display.pack(fill=tk.BOTH, expand=True)


user_input = tk.Entry(window)
user_input.pack(fill=tk.X)

send_button = tk.Button(window, text="Send", command=send_message)
send_button.pack()


custom_question_label = tk.Label(window, text="Custom Question:")
custom_question_label.pack()
custom_question_entry = tk.Entry(window)
custom_question_entry.pack(fill=tk.X)

custom_answer_label = tk.Label(window, text="Custom Answer:")
custom_answer_label.pack()
custom_answer_entry = tk.Entry(window)
custom_answer_entry.pack(fill=tk.X)


add_custom_button = tk.Button(window, text="Add Custom Q&A", command=add_custom_qa)
add_custom_button.pack()

window.bind("<Return>", lambda event=None: send_message())


window.mainloop()
```
